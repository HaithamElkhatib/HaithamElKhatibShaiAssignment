### Task1: **Basic Data Exploration**

```python
import pandas as pd
import numpy as np

# Load your dataset
df = pd.read_csv('/content/Salaries.csv')
df.head()
```

```python
# Get the number of rows and columns
number_rows = df.shape[0]
number_columns = df.shape[1]

# Determine the data types of each column
data_types = df.dtypes

# Check for missing values in each column
missing_values = df.isnull().sum()

# Print the results
print(f"Number of rows: {number_rows}\n")
print(f"Number of columns: {number_columns}\n")
print(f"Data types: {data_types}\n")
print(f"Missing values: {missing_values}")
```

### Task2: **Descriptive Statistics**

```python
# Calculate basic statistics
mean_salary = df['TotalPay'].mean()
median_salary = df['TotalPay'].median()
mode_salary = df['TotalPay'].mode()[0]
minimum_salary = df['TotalPay'].min()
maximum_salary = df['TotalPay'].max()

# Calculate range
salary_range = maximum_salary - minimum_salary

# Calculate standard deviation
salary_std = df['TotalPay'].std()

# Print the results
print(f"Mean salary: {mean_salary}\n")
print(f"Median salary: {median_salary}\n")
print(f"Mode salary: {mode_salary}\n")
print(f"Minimum salary: {minimum_salary}\n")
print(f"Maximum salary: {maximum_salary}\n")
print(f"Salary range: {salary_range}\n")
print(f"Salary standard deviation: {salary_std}\n")
```

### Task3: **Data Cleaning**


```python
# Impute missing values with the median
df['BasePay'].fillna(df['BasePay'].median(), inplace=True)
df['OvertimePay'].fillna(df['OvertimePay'].median(), inplace=True)
df['OtherPay'].fillna(df['OtherPay'].median(), inplace=True)
df['Benefits'].fillna(df['Benefits'].median(), inplace=True)
# Drop columns with a large number of missing values
df.drop(['Status', 'Notes'], axis=1, inplace=True)



# Check for missing values in each column
missing_values = df.isnull().sum()
print(f"Missing values: {missing_values}")
```

### Task4: **Basic Data Visualization**


```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Histogram for BasePay
plt.hist(df['BasePay'])
plt.xlabel('BasePay')
plt.ylabel('Frequency')
plt.title('Distribution of BasePay')
plt.show()

# Boxplot for OvertimePay
plt.boxplot(df['OvertimePay'])
plt.xlabel('OvertimePay')
plt.ylabel('Amount')
plt.title('Distribution of OvertimePay')
plt.show()

# Scatter plot for BasePay vs. TotalPay
plt.scatter(df['BasePay'], df['TotalPay'])
plt.xlabel('BasePay')
plt.ylabel('TotalPay')
plt.title('Relationship between BasePay and TotalPay')
plt.show()

# Boxplot for BasePay by JobTitle
sns.boxplot(x=df['JobTitle'], y=df['BasePay'])
plt.xlabel('JobTitle')
plt.ylabel('BasePay')
plt.title('Distribution of BasePay by JobTitle')
plt.show()

# Bar chart for JobTitle
plt.bar(df['JobTitle'].value_counts().index, df['JobTitle'].value_counts().values)
plt.xlabel('JobTitle')
plt.ylabel('Frequency')
plt.title('Frequency of Job Titles')
plt.show()

# Pie chart for Agency
plt.pie(df['Agency'].value_counts().values, labels=df['Agency'].value_counts().index, autopct='%1.1f%%')
plt.title('Proportion of Employees by Agency')
plt.show()
```

### Task5: **Grouped Analysis**


```python
# Group data by JobTitle
grouped_data = df.groupby('JobTitle')

# Calculate summary statistics for each group
summary_stats = grouped_data.describe()

# Print the summary statistics
print(summary_stats)
```

```python
average_salaries = summary_stats['BasePay']['mean']

# Print the average salaries for each JobTitle
print(average_salaries)
```

```python
grouped_data = df.groupby(['JobTitle', 'Agency'])

# Calculate summary statistics for each group
summary_stats = grouped_data.describe()

# Print the summary statistics
print(summary_stats)
```

### Task6: **Simple Correlation Analysis**


```python
# Calculate the correlation coefficient
correlation = df['BasePay'].corr(df['OvertimePay'])

# Print the correlation coefficient
print(f"Correlation between BasePay and OvertimePay: {correlation}")

# Plot the scatter plot
plt.scatter(df['BasePay'], df['OvertimePay'])
plt.xlabel('BasePay')
plt.ylabel('OvertimePay')
plt.title('Scatter Plot of BasePay vs. OvertimePay')
plt.show()
```
